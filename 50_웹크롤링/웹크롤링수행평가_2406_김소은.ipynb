# 2021년 8월 국내도서 종합 월별 베스트셀러 TOP40

* YES24 사이트
* http://www.yes24.com/main/default.aspx
* 책 제목, 저자, 출간일, 현재 가격 추출

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'http://www.yes24.com/24/category/bestseller?CategoryNumber=001&sumgb=09&year=2021&month=8&FetchSize=40'

headers = {
   'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/93.0.4577.63 Safari/537.36'
}

res = requests.get(url, headers=headers)
res
```

```python
res.text[:300]
```

```python
soup = BeautifulSoup(res.text)
soup
```

### 책 제목 추출

```python
books = soup.select('.goodsTxtInfo > p:first-child > a:first-child')
title = []

for book in books:
    title.append(book.get_text().strip())
    
title
```

### 저자 추출

```python
books_first = soup.select_one('.goodsTxtInfo > div > a:nth-child(2)')
books = soup.select('.goodsTxtInfo > div > a:first-child')[1:]
author = []

author.append(books_first.get_text())
for book in books:
    author.append(book.get_text())
    
author
```

### 출간일 추출

```python
books = soup.select('.goodsTxtInfo > div')
release = []

for book in books:
    release.append(book.get_text().split('|')[2].strip())
    
release
```

### 현재 가격 추출

```python
books = soup.select('.goodsTxtInfo > p > .priceB')
cost = []

for book in books:
    cost.append(book.get_text())
    
cost
```

### 책 제목, 저자, 출간일, 현재 가격 dataframe 생성

```python
df_data = {
    'title': title,
    'author': author,
    'release': release,
    'cost': cost
}

df = pd.DataFrame(data=df_data)
df
```

```python
temp = xmltodict.parse(res.text)
temp
```

```python
import json
result = json.loads(json.dumps(temp))
result
```

```python
items = result['response']['body']['items']['item']
so2Value = []  # 아황산가스 농도
coValue = []  # 일산화탄소 농도
o3Value = []  # 오존 농도
no2Value = []  # 이산화질소 농도
pm10Value = []  # 미세먼지 농도

for item in items:
    so2Value.append(item['so2Value'])
    coValue.append(item['coValue'])
    o3Value.append(item['o3Value'])
    no2Value.append(item['no2Value'])
    pm10Value.append(item['pm10Value'])
```

```python
df_data = {
    '아황산가스 농도': so2Value,
    '일산화탄소 농도': coValue,
    '오존 농도': o3Value,
    '이산화질소 농도': no2Value,
    '미세먼지 농도': pm10Value
}
df_data
```

```python
import pandas as pd
df = pd.DataFrame(data=df_data)
df
```
